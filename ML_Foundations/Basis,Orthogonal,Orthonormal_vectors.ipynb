## Basis, Orthogonal, Orthonormal_vectors

### Start with the Basis Vector

### " SIMPLY THE BASIS VECTOR IS KIND OF VECTOR WHICH CAN BE SCALED TO REPRESENT ANY VECTOR IN A GIVEN VECTOR SPACE "

```python
import numpy as np
```

```python
# 1. Defining the standard basis Vector : 
i = np.array([1,0])
j = np.array([0,1])

# 2. Define the scalars (the "stretching" amounts)
stretch_scalar_i = 5
stretch_scalar_j = 2.5
# 3. Perform the linear combination (scaling and adding)
# This is exactly the operation: V = 5*i + 2.5*j

V = (stretch_scalar_i*i) + (stretch_scalar_j*j)
V
```

## Orhtogonal Vector ------
## Orthogonal Vector is just dot product of any vector (it could be any dimention)
## * Orhtogonal Vectors Value is null and they are 90 degree to each other--

```python
# Example ----
i = np.array([1,0,0])
j = np.array([0,1,0])
k = np.array([0,0,1])

# Dot product of these vector will be like - (i.j) + (j.k) + (i.k)
np.dot(i,j)
np.dot(j,k)
np.dot(i,k)
```

## Orthonormal Vector ------
## Orthonormal vector is the same as orthogonal vector but a unit vector (Euclidien Distance)

```python
# Example ----
i = np.array([1,0,0])
j = np.array([0,1,0])
k = np.array([0,0,1])

# First Criteria : They must have to be orthogonal

# Dot product of these vector will be like - (i.j) + (j.k) + (i.k)
np.dot(i,j)
np.dot(j,k)
np.dot(i,k)
```

```python
# Second Criteria: They also must have to be unit vector like : x_axis=[1,0,0] , y_axis=[0,1,0] , z_axis=[0,0,1]

L2_norm_i = ((i[0]**2) + (i[1]**2) + (i[2]**2)) ** 0.5
L2_norm_j = ((j[0]**2) + (j[1]**2) + (j[2]**2)) ** 0.5
L2_norm_k = ((k[0]**2) + (k[1]**2) + (k[2]**2)) ** 0.5

L2_norm_i, L2_norm_j, L2_norm_k
```
