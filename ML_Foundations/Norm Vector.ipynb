```python
n1 = 4
n2 = 12
n3 = 20
```

## L2 norm is just a fancy way of saying the distance. Reffered as : ||x||2   or ||x||

```python
## Lets say (x1, x2) = (4,12)
## So L2 norm is : 
L2_norm = (4**2 + 12**2)*(1/2)
L2_norm
```

## L1 norm is just simply adding the values of a vector { Absolute Values }. Refferd as: ||x||1

```python
x1 = -4
x2 = -12
x3 = 20

# Let's say V = [x1 = -4, x2 = -12, x3 = 20] 
```

```python
L1_norm = np.abs(x1) + np.abs(x2) + np.abs(x3)
L1_norm
```

```python
## Squared L2 norm is just adding those values but squared. Denoted as : ||x||2
##                                                                            2
```

```python
squared_l2_norm = (n1**2) + (n2**2) + (n3**2)
squared_l2_norm
```

## Also squared L2 norm = X^T.X

```python
x = np.array([[4,12,20]])
x
```

```python
x_transposed = x.T
x_transposed
```

```python
s_l2_norm = x_transposed*x
s_l2_norm
```

## Max Norm
### Max norm means the maximum absolute value of a vector

```python
Mx_nm = [-25,-20,5]

max_norm = np.max([np.abs(-25),np.abs(-20),np.abs(5)])
max_norm
```
