## Vectors (Rank 1 Tensors) in NumPy

### Vectors and Vector Transposition
### In General Vector Transposition means Column and Rows Transposed Their Gender
### Column -> Rows, Rows -> Column

```python
# Vector Transposition ------
import numpy as np
x = np.array([1,2,3,4,5,6])
x
```

```python
# Transposing a regular 1-D array has no effect...
x_trans = x.T
x_trans
```

```python
x_trans.shape
```

```python
x_trans.ndim
```

```python
# ...but it does we use nested "matrix-style" brackets:
x_ms = np.array([[1,2,3,4]])
x_ms.T
```
