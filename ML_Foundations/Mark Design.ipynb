```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0,100,1000) # Start,Stop,n points
```

### Energy Generated By Mark1 :  e = 1*time

```python
mark1_energy = 1*time
```

### Energy Generated By Mark2: e = 4*(time-30)

```python
mark2_energy = 4*(time - 30)
```

```python
# Create a figure and a set of subplots (fig is the figure object, ax is the axes object)
fig,axis = plt.subplots()
# Set the title of the plot to "Equal Energy Generate"
plt.title("Equal Energy Generate")
# Label the x-axis as "Time (In Day)"
plt.xlabel("Time (In Day)")
# Label the y-axis as "Power (In KJ)"
plt.ylabel("Power (In KJ)")
# Set the x-axis limits from 0 to 60 minutes
axis.set_xlim([0,60])
# Set the y-axis limits from 0 to 120 minutes
axis.set_ylim([0,120])
# Plot the Mark1 energy over time in green
axis.plot(time, mark1_energy, color="green")
# Plot the Mark2 energy over time in black
axis.plot(time, mark2_energy, color="black")
# Draw a vertical dashed blue line at x = 40 (day when both of the machine's energy meets each other)
plt.axvline(x=40, color="blue", linestyle="--")
# Draw a horizontal dashed blue line at y = 80 (energy scale when both of the machine's energy meets each other)
plt.axhline(y=80, color="blue", linestyle="--")
```
