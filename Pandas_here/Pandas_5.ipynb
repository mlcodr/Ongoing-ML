```python
car_sales_missing = pd.read_csv("D:\Olympiad\CSV FILES\car-sales-missing-data.csv")
car_sales_missing
```

```python
## Assigning the mean value of all the odometer.
car_sales_missing['Odometer'] = car_sales_missing['Odometer'].fillna(car_sales_missing['Odometer'].mean())
```

```python
## now let's see the missing dataframe
car_sales_missing
```

```python
# .dropna() is used to remove all of the rows which contains NaN value
c_s_m = car_sales_missing.dropna(inplace=False)
```

```python
# Creating a column using pandas series for concatinate with the dataframe
seats_column = pd.Series([3,3,3,3,3,3,3,3,3,3])
car_sales["Seats"] = seats_column
car_sales
```

```python
Fuel_Tank = pd.Series([10,8,12,14,16,18,12,20,30,40])
car_sales['Fuel (Litre)'] = Fuel_Tank
car_sales
```

```python
# Create a column from a Python list
engine_sizes = [1.3, 2.0, 3.0, 4.2, 1.6, 1, 2.0, 2.3, 2.0, 3.0]
car_sales["Engine Size"] = engine_sizes
car_sales
```

```python
## Error says Price column is String , so we have to convert it to integer
car_sales['Price'] = car_sales['Price'].str.replace("[\$,]" , "", regex=True).astype(float)
```

```python
car_sales.info()
```

```python
car_sales['Price_per_KM'] = car_sales['Odometer (KM)'] / car_sales['Price']
car_sales
```

```python
car_sales["Wheel No"] = 4
car_sales
```

```python
car_sales["Passed Road Safety"] = True
car_sales
```

```python
# Wheel no column looks unusual, lets remove it.
car_sales.drop("Wheel No",axis=1)
```
