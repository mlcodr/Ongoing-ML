## Datatypes in Pandas

```python
# Creating a series Product
Product_name = pd.Series(["Headset","Earbud","Headphone"])
Product_name
```

```python
Product_Price = pd.Series(["5$","8$","7$"])
Product_Price
```

```python
# Creating a DataFrame of cars and colours
car_data = pd.DataFrame({"Product": Product_name, 
                         "Price": Product_Price})
car_data
```

```python
foods = pd.Series(["Cold Coffee","Chicken Fries","Banana","Dates","Milk"])
```

```python
prices = pd.Series(["5$","4$","7$","9$","6$"])
```

```python
food_df = pd.DataFrame({
    "Items":foods,
    "Price":prices
})
```
