```python
# Legend = The box labeled Legend (Target 0/1). It explains what the colors or markers represent.
```

```python
fig,(axes_5,axes_6) = plt.subplots(nrows=2,ncols=2,figsize=(10,8))
fig.suptitle("Butter Prices by Years")
for ax in axes_5:
    ax.plot([20,21,22],[35,44,30],color = "blue",label = "Price Chart")
    ax.set_title("Cashew Butter")
    ax.set_xlabel("Times")
    ax.set_ylabel("Price")
    ax.legend()
for ax in axes_6:
    ax.plot([24,25,26],[40,50,45],color = "Yellow",label = "Price Chart")
    ax.set_title("Milk Butter")
    ax.set_xlabel("Times")
    ax.set_ylabel("Price")
    ax.legend()
plt.tight_layout()
```

```python
# Generating Dummy Data 
np.random.seed(0)
age = np.random.randint(50,80,size=100)
cholesterol = np.random.randint(150,400,size=100)
max_heart_rate = np.random.randint(80,130,size=100)
target = np.random.choice([0,1],size=100) # 0 = No Disease, 1 = Disease 

# Creating Figure and Axes or (canvas and subplot)
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize = (10,8),sharex=True)
fig.suptitle("Heart Disease Analysis",fontsize = 20, fontweight = "bold", color = "Navy")

# Plotting the first axes (ax0)
scatter0 = ax0.scatter(age,cholesterol,c = target ,cmap= "winter", alpha=1)
ax0.set_xlabel("Age")
ax0.set_title("Heart Disease and Cholesterol Levels")
ax0.set_xlabel("Cholesterol")
ax0.axhline(y=250,color="r",linestyle = "--")
ax0.grid(True)

# Plotting the second axes (ax1)
scatter1 = ax1.scatter(age,max_heart_rate,c=target,cmap="winter",alpha=1)
ax1.set_title("Heart Disease and Max Heart Rate Levels")
ax1.set_xlabel("Age")
ax1.set_ylabel("Cholesterol")
ax1.axhline(y=140,color="r",linestyle="--")
ax1.grid(True)

# Adding the Legend
legend_elements = scatter1.legend_elements()
ax1.legend(*legend_elements,title="Target")

# Adjusting the Layout
plt.tight_layout();
```
