# Bar Plots from a pandas DataFrame.

```python
x = np.random.rand(10,4)
df1 = pd.DataFrame(x,columns=["W","X","Y","Z"])
df1
```

```python
# We can Turn the dataframe into a plot by just using plot.bar()
df1.plot.bar()
plt.grid(True)
```

```python
# Same thing in another way 
df1.plot(kind = "bar")
plt.grid(True)
plt.xlabel("Order Of Column")
plt.ylabel("Values")
plt.suptitle("Bar Plot");
```

```python
# Create Figure and Axes instance
fig, ax = plt.subplots(figsize=(10, 6))

# Plot data directly to the Axes intance
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"]) # Colour the data with the "target" column

# Customize the plot parameters 
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Setup the legend
ax.legend(*scatter.legend_elements(), 
          title="Target");
```
