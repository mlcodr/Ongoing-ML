## Comprehensive plotting by instance axes. 

```python
# Creating the figure
fig,ax = plt.subplots(figsize = (10,8))
# Plotting the data to the axes
scatter = ax.scatter(heart_disease["age"], 
                    heart_disease["chol"], 
                     c = heart_disease["target"])
# Customizing the plot 
ax.set(title="Heart Disease and Cholesterol Levels", 
      xlabel="Age", 
      ylabel="Cholesterol")
# Setting up the legend
ax.legend(*scatter.legend_elements(),
         title="Target");
# adding the meanline 
ax.axhline(heart_disease["chol"].mean(),
          linestyle = "--");
# adding the vertical line 
ax.axvline(heart_disease["age"].mean(), 
          linestyle = "--");
```

```python
heart_disease
```

```python
# First We need to make the figure 
# Secondly we need to put our data in the figure 
# Thirdly We need to customize our data as we want 
# Fourth we need to add a mean line (Average)
# If need adding a legend will be sweet !!!!!! 

                                      # Do this for both of the axes 

fig , (axes_1,axes_2) = plt.subplots(figsize = (10,8), nrows=2, ncols= 1 , sharex = True)

scatter = axes_1.scatter(heart_disease["age"],
                        heart_disease["chol"],
                        c = heart_disease["target"])

axes_1.set(title = "Heart Disease & Cholesterol Level",
          xlabel = "Age", 
          ylabel = "Cholesterol")

axes_1.axhline( y = heart_disease["chol"].mean(), 
                linestyle = "--",
                label = "Average", 
                color = "gray")

axes_1.legend(*scatter.legend_elements(),
             title = "Target")
# **********************************************************************************************************
scatter = axes_2.scatter(heart_disease["age"],
                        heart_disease["thalach"],
                        c = heart_disease["target"])

axes_2.set(title = "Heart Disease & Max Heart Level",
          xlabel = "Age", 
          ylabel = "Max Heart Rate")

axes_2.axhline( y = heart_disease["thalach"].mean(), 
                linestyle = "--",
                label = "Average", 
                color = "gray")

axes_2.legend(*scatter.legend_elements(),
             title = "Target")


plt.suptitle("Heart Disease Analysis",
            fontsize = 16,
            fontweight = "bold")
```

```python
## Customizing the colours of plot with colourmaps (cmap)
fig,ax = plt.subplots(figsize = (8,6), nrows = 1 , ncols = 1)
scatter_plot = ax.scatter(heart_disease['age'], 
                         heart_disease['chol'], 
                         c = heart_disease["target"],
                         cmap = "viridis")  ## Default Cmap Value

ax.set(title = "Colour Map Watching",
      xlabel = "Age", 
      ylabel = "Cholesterol",
      xlim = (30,100))

ax.legend(*scatter_plot.legend_elements(),
         title = "Target",
         loc = "upper left")

ax.axhline( y = heart_disease["chol"].mean(),
          label = "Average",
          linestyle = "--",
          c = "gray");
```
