# Bar Plots from a pandas DataFrame.

```python
x = np.random.rand(10,4)
df1 = pd.DataFrame(x,columns=["W","X","Y","Z"])
df1
```

```python
# We can Turn the dataframe into a plot by just using plot.bar()
df1.plot.bar()
plt.grid(True)
```

```python
# Same thing in another way 
df1.plot(kind = "bar")
plt.grid(True)
plt.xlabel("Order Of Column")
plt.ylabel("Values")
plt.title("Bar Plot");
```

```python
# Let's try bar plot on a DataFrame this time we will specify x and y axis values 
car_sales.plot(x = "Make", y = "Price", kind = "bar", color = "skyblue")
plt.title = "Car Sales"
plt.grid(True)
```
