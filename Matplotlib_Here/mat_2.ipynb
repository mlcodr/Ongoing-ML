# A quick Matplotlib Workflow ---

```python
# 1. Prepare Data 
x = [1,2,3,4]
y = [11,20,41,35]
# 2. Setup Plot 
fig,axes = plt.subplots(figsize = (10,8))
# 3. Plotting Data 
axes.plot(x,y)
# 4. Customize The Plot
axes.set(
    title = "Sample Plot",
    xlabel = "X-Axis",
    ylabel = "Y-Axis"
);
```

```python
# Creating an array 
x = np.linspace(10,20,50)  # Linspace(Start:Stop:Number of elements) is used to generate random float numbers which is linear and evenly spaced.
x[:10]

# Creating Figure and Axes 
fig,axes = plt.subplots()
axes.plot(x,x**2.5)
```

```python
## Defined Data
days = ['Sat','Sun','Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales_last_week = [10, 15, 12, 25, 20,13,18]
sales_this_week = [12, 18, 14, 28, 35,37,45]
# Setup Plot
fig,ax = plt.subplots()
# Customizing Plots
ax.plot(days,sales_last_week,color = "red",marker="*",linestyle = "--",label = "Last Week")
ax.plot(days,sales_this_week,color = "Blue", marker = "o",label = "This Week") # These title are for legend
ax.set(ylabel="Sales($)",
      title = "Weekly Sales Comparison")
ax.legend()
ax.grid(True);
```
