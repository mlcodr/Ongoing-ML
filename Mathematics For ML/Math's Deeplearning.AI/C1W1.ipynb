```python
# Stacking 
# Stacking is the feature of Numpy that leads to increased customizaiton of arrays.
# It means to join two or more arrays, either horizontally or vertically. Meaning that it's done along a new axis
```

```python
# np.vstack() - stack vertically 
# np.hstack() - stack horizontally 
# np.hsplit() - splits an array into several smaller arrays. 
```

```python
a1 = np.random.randint(8,12,size=(2,4))
a2 = np.random.randint(10,14,size=(2,4))
a1,a2
```

```python
vert_arr = np.vstack((a1,a2))
vert_arr   ## It glues two arrays vertically
```

```python
hor_arr = np.hstack((a1,a2))
hor_arr   ## It glues arrays horizontally
```
