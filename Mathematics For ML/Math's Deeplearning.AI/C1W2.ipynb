# Solving Linear Systems: 3 Variables

## Representing and Solving a system of Linear Equations using Matrices

#### System of Linear System with three equaiton and three unknown variables. 
$$
\begin{cases}
4x_1 - 3x_2 + x_3 = -10, \\
2x_1 + x_2 + 3x_3 = 0, \\
-x_1 + 2x_2 - 5x_3 = 17
\end{cases}
$$

```python
## Seperating Co-effieciant and Values 
A = np.array([[4,-3,1],[2,1,3],[-1,2,-5]])
B = np.array([-10,0,17])
```

## Solving system of Linear Equations using Row Reduction 
#### Functions for Elementary Operations 
#### Let's review *elementary oprations*, which don't change the solution set of a linear system: 
* Multiply any row by a non-zero number
* Add two rows and exchange one of the original rows with the result of the addition
* Swap rows
#### In the case of larger systems you will need to apply elementary operations multiple time,
#### Thus it's convenient to define the correspoding pyhton functions.
#### Investigate them in the following code cells with the example

```python
## Preperation for Row Reduction ------
# Reshaping B 
B = B.reshape(3,1)
## Unifying 
aug_mat = np.hstack((A,B))
aug_mat
```

```python
## Multiplying any row by a non zero number 
def multiplyrow(matrix,row_index,multiplier):
    
    ## Copy of main matrix 
    matrix_new = matrix.copy()

    ## Main Equation 
    matrix_new[row_index] = matrix_new[row_index] * multiplier

    return matrix_new

print(f"Prime Matrix : {aug_mat}")
print(f"Prime Matrix after multiplying third row by 2 : {multiplyrow(aug_mat,2,2)}")
```

```python
## Add two rows and exchange one of the original rows with the result of the addition
def addrow(matrix,row_index_first,multiplier,row_index_second,):
    
    ## Copy of main matrix 
    matrix_new = matrix.copy()

    ## Main Equation 
    matrix_new[row_index_second] = matrix_new[row_index_first] * multiplier + matrix_new[row_index_second]

    return matrix_new

print(f"Prime Matrix : {aug_mat}")
print(f"Prime Matrix after adding two rows and exchange one of the original rows : {addrow(aug_mat,1,1/2,2)}")
 
```

```python
# exchange row_index_first and row_index_second of the matrix M
def swaprows(matrix,row_index_first,row_index_second): 

    ## Copy of main matrix 
    matrix_new = matrix.copy()

    ## Main Equation 
    matrix_new[[row_index_first,row_index_second]] = matrix_new[[row_index_second,row_index_first]]

    return matrix_new

print(f"Prime Matrix : {aug_mat}")
print(f"Prime Matrix after swapping first and 3 rows: {swaprows(aug_mat,0,2)}")
```

## Row Reduction and Solution of the Linear System 
#### Now you can use the defined operations to bring the matrix into row reduced form.
#### To do this manually, it it convenient to have 1 or -1 value in the first element 
#### of the first row(the arihtmetics of operations is easier then).
#### Performing calculations in python, won't provide much of a benefit,
#### but it is better to do that for illustration purposes, So, Let's swap the first and third rows: 


```python
# note : ref is and abbreviation of the row echelon form (row reduced form ) 
A_ref = swaprows(aug_mat,0,2)
A_ref
```

```python
## To turn the first element of the first and second row to zero we need to multiply it by 2 and add to second row 
A_ref = addrow(A_ref,0,2,1)
A_ref
```

```python
A_ref = addrow(A_ref,0,4,2)
A_ref
```

```python
# The next step will be to turn the second element of the third row to zero
A_ref = addrow(A_ref,1,-1,2)
A_ref
```

```python
# Now its easy to find out the value of x3 from the third row as it is: -12x_3 = 24
# Divide third row by -12 
A_ref = multiplyrow(A_ref,2,-1/12)
A_ref
```

```python
## now x3 = -2 , so x2 and x1 will be 
x_3 = -2
x_2 = (A_ref[1,3] - A_ref[1,2] * x_3) / A_ref[1,1]
x_1 = (A_ref[0,3] - A_ref[0,2] * x_3 - A_ref[0,1] * x_2) / A_ref[0,0]
print(f"Value of X1,X2,X3 = {x_1},{x_2},{x_3}")
```

##  System of Linear Equations with infinite numbers of solution
