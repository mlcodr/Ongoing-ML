```python
                                        ## Array math operations 
arr_1 = np.arange(1,10,2)
arr_2 = np.arange(1,20,4)
# Basic operation between two 1-d arrays 
addition = arr_1 + arr_2 ,
subtraction = arr_1 - arr_2 ,
multiplication = arr_1 * arr_2 ,
division = arr_2 / arr_1
```
