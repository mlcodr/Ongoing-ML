The following equation should be :
$$
\begin {cases}
7f + 5a + 3c = 120, \\
3f + 2a + 5c = 70, \\
f + 2a + a = 20
\end {cases}
$$
note : f for fiberglass, a for aluminium, c for carbon 

### The following steps that can be taken to solve this linear equation. 
* Muliply by a scalar and add two rows
* Multiply the first equation by 3 and substract it from equation 2
* Divide the first equation by 7
* Substract the second row from the first row
* Isolate one variable and substitute into the next eqaution to find the other variables
  

### These are the following information that can be extracted from the equation. 
* Whether the matrix is singular or non-singular
* The rank of the matrix
* Row-Reduced echelon form
* Number of linearly (dependent or independent) rows and columns

### This matrices represents the system for all three iteration

```python
A = np.array([[7,5,3],[3,2,5],[1,2,1]])
A
```

```python
### Calculation the cost of each materials by solving the system of equations 
b = np.array([120,70,20])
b = b.reshape(3,1)
A_system = np.hstack((A,b))
A_system
```

```python
cost = np.linalg.solve(A,b).astype(int)
f,a,c = cost
print(f"The cost of each materials is following {f}$ , {a}$, {c}$")
```

```python
## Using determinant to find out if the matrix is singular or non-singular 
det_A = np.linalg.det(A)
if np.isclose(det_A,0.00):
    print(f"The Determinant is {det_A:.2f}. The matrix is singular")
else: 
    print(f"The Determinant is {det_A:.2f}. The matrix is Non-Singular")
```

```python
## The rank in the above matrix is
np.linalg.matrix_rank(A)
```

 Here "b" comes first because already one row is solved 
 "a" comes second because if row number 3 get multiplied by -2, then its easy to find out the value 
 "c" comes last. There is no simple way to multiply one row to get another, and no row is a sum of the others.
