## Practice quiz - solving system of linear equations 

## The following system of equation will be : 
#### (b) for basalt rock and (m) for meteorite rock
$$
\begin {cases}
2b + 3m = 15, \\
2b + 4m = 16
\end {cases}
$$

```python
## Following system of linear eqaution represents this matrices 
## Seperating the co - efficiant and values 
A = np.array([[2,3],[2,4]])
B = np.array([15,16])
A,B
```

```python
## Determinant of the matrices 
det_A = np.linalg.det(A)
det_A
```

### It has value which is ( not=0 ) , That means -----
### It's a non singular matrix 
### It has Linearly Dependent Rows 
### It has a numerical value of b and m 

```python
## Solving it
rock_weight = np.linalg.solve(A,B)
b,m = rock_weight
print(f"Weight of Basalt & Meteorite rock is : {b}g, {m}g")
```

## Graded Quizes

## The following system of equation should be : 
#### (b) for basalt rock and (m) for meteorite rock & (c) for crystal rock
$$
\begin {cases}
2b + 1m + 5c = 20, \\
b + 2m + c = 10, \\
2b + m + 3c = 15
\end {cases}
$$

```python
## Following system of linear eqaution represents this matrices 
## Seperating the co - efficiant and values 
A = np.array([[2,1,5],[1,2,1],[2,1,3]])
B = np.array([20,10,15])
A,B
```

```python
## Determinant of the matrices 
det_A = np.linalg.det(A)
det_A
```

### It has value which is ( not=0 ) , That means -----
### It's a non singular matrix 
### It has Linearly Dependent Rows 
### It has a numerical value of b and m and c

```python
## Solving it
rock_weight = np.linalg.solve(A,B)
b,m,c = rock_weight
print(f"Weight of Basalt & Meteorite & Crystal rock is : {b}g, {m}g, {c}g")
```
