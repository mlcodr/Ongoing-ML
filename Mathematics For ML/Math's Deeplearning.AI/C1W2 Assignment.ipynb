### Systems of Linear Equations and Corresponding Numpy Arrays

#### Equation
$$
\begin {cases}
2x1 - x2 + x3 + x4 = 6, \\
x1 + 2x2 - x3 - x4 =3, \\
-x1 + 2x2 + 2x3 + 2x4 = 14, \\
x1 - x2 + 2x3 + x4 = 8
\end {cases}
$$

```python
## Constructing A and b
A = np.array([[2,-1,1,1],[1,2,-1,-1],[-1,2,2,2],[1,-1,2,1]],dtype = np.dtype(int))
b = np.array([6,3,14,8],dtype = np.dtype(int))
```

### Solution of Equation using Linear Algebra Package 

```python
# Determinant of matrix A 
det_A = np.linalg.det(A)
print(f"Determinant of the matrix A is: {det_A:.2f}, Solution Vector is: {np.linalg.solve(A,b)}")
```

### Elementary operations and Row Reductions
Elementary operations can be done by three simple steps 
* Multiply any row by non-zero number
* Add two rows and exchange one of the original rows with the result of the addition
* Swapping rows

* *In order to perform these operations we need to create three functions that will be used frequently to solve the Linear Equation*
*  **multiplyrows, addrows, swaprows**

```python
def multiplyrows(matrix,row_index,multiplier): 
    new_matrix = matrix.copy()
    new_matrix[row_index] = new_matrix[row_index] * multiplier
    return new_matrix
```

```python
def addrows(matrix,row_num_one,multiplier,row_num_two):
    new_matrix = matrix.copy()
    new_matrix[row_num_two] = new_matrix[row_num_one] * multiplier + new_matrix[row_num_two]
    return new_matrix
```

```python
def swaprows(matrix,row_num_one,row_num_two):
    new_matrix = matrix.copy()
    new_matrix[[row_num_one,row_num_two]] = new_matrix[[row_num_two,row_num_one]]
    return new_matrix
```

```python
 # stack horizontally matrix A and vector b, which needs to be reshaped as a vector (4, 1)
b = b.reshape(4,1)
A_system = np.hstack((A,b))
A_system
```

```python
# swap row 0 and row 1 of matrix A_system (remember that indexing in NumPy array starts from 0)
A_ref = swaprows(A_system,0,1)
A_ref
```

```python
# multiply row 0 of the new matrix A_ref by -2 and add it to the row 1
A_ref = addrows(A_system,0,-2,1)
A_ref
```

```python
# multiply row 0 of the new matrix A_ref by -2 
A_ref = multiplyrows(A_system,0,-2)
A_ref
```
