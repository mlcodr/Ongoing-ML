```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0,40,1000) # Start,Finish, n points
```

### Distance Travelled By The Robber: d = 2.5t

```python
distance_robber = 2.5*time
```

### Distance Travelled By The Sheriff: d = 3(t-5)

```python
distance_sheriff = 3*(time - 5)
```

```python
# Create a figure and a set of subplots (fig is the figure object, ax is the axes object)
fig, ax = plt.subplots()

# Set the title of the plot to "Bank Robber Caught"
plt.title("Bank Robber Caught")

# Label the x-axis as "Time (In Minutes)"
plt.xlabel("Time (In Minutes)")

# Label the y-axis as "Distance (In Kilometre)"
plt.ylabel("Distance (In Kilometre)")

# Set the x-axis limits from 0 to 40 minutes
ax.set_xlim([0, 40])

# Set the y-axis limits from 0 to 100 kilometres
ax.set_ylim([0, 100])

# Plot the robber's distance over time in green
ax.plot(time, distance_robber, color="green")

# Plot the sheriff's distance over time in black
ax.plot(time, distance_sheriff, color="black")

# Draw a vertical dashed purple line at x = 30 (time when robber was caught)
plt.axvline(x=30, color="purple", linestyle="--")

# Draw a horizontal dashed purple line at y = 75 (distance where robber was caught)
plt.axhline(y=75, color="purple", linestyle="--")
```
