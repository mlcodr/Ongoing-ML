```python
import numpy as np
import matplotlib.pyplot as plt

# Time range
time = np.linspace(0, 100, 100)

# Energy Generated
mark1_energy = 1 * time
mark2_energy = 4 * (time - 30)

# Create figure and subplots
fig, axs = plt.subplots(2, figsize=(10, 6))
fig.suptitle("Figure of Equal Energy Generated")

# Mark1 Plot
axs[0].set_title("Mark1 Energy")
axs[0].set_xlabel("Time (in Days)")
axs[0].set_ylabel("Power (in K)")
axs[0].plot(time, mark1_energy, color="green")
axs[0].axhline(y=80, color="black", linestyle="--", label="Threshold = 80K")
axs[0].axvline(x=80, color="red", linestyle=":", label="Mark1 hits 80K")
axs[0].grid(True)
axs[0].legend()

# Mark2 Plot
axs[1].set_title("Mark2 Energy")
axs[1].set_xlabel("Time (in Days)")
axs[1].set_ylabel("Power (in K)")
axs[1].plot(time, mark2_energy, color="green")
axs[1].axhline(y=80, color="black", linestyle="--", label="Threshold = 80K")
axs[1].axvline(x=50, color="red", linestyle=":", label="Mark2 hits 80K")
axs[1].grid(True)
axs[1].legend()

# Common horizontal line at y = 80 across both plots
plt.axhline(y=80, color="blue", linestyle="--")

plt.tight_layout()
plt.show()
```
