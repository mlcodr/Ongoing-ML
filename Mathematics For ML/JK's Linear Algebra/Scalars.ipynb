## Scalar, Vector, Matrices and Tensor
### Scalar - 2,5,-8,9,0.5 (A single numerical value) - 0 dimentional
### Vector - [2, 4, -1] (An ordered list of numbers (magnitude and direction)) - 1 dimentional
### Matrices - [[1, 2],[3, 4]] (A 2D array of numbers arranged in rows and columns) - 2 dimentional
### Tensors - [[[1,2],[3,4]],[[5,6],[7,8]]] (A generalization of scalars, vectors, and matrices to higher dimensions) - 3,4,5,6....n dimentional

### tf.Variable
### tf.constant

```python
x_tf = tf.Variable(95,dtype = tf.int32)
x_tf
```

```python
y_tf = tf.Variable(5,size=(tf.int32))
y_tf
```

```python
x_tf.shape
```

```python
sum_tf = tf.add(x_tf,y_tf)
sum_tf
```

```python
sub_tf = tf.subtract(sum_tf,y_tf)
sub_tf
```

```python
sum_tf.numpy()   # note that NumPy operations automatically convert tensors to NumPy arrays, and vice versa
```

```python
type(sum_tf)
```

```python
sum_tf.dtype
```
