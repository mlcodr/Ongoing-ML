# Nut Butter Sales (Dot Product Example)

```python
import numpy as np
import pandas as pd 
```

```python
np.random.seed(0)
# Numbers of jars sold
sales_amount = np.random.randint(0,20,size=(6,3))
sales_amount
```

```python
# Creating Weekly Sales Dataframe
weekly_sales = pd.DataFrame(sales_amount,index=["Sat","Sun","Mon","Tues","Wed","Thur"],
                            columns=["Almond Butter","Peanut Butter","Cashew Butter"])
weekly_sales
```

```python
# Creating Prices array
prices = np.array([12,10,14])
prices
```

```python
# Creating Prices Dataframe
price_df = pd.DataFrame(prices.reshape(1,3),index = ["Price"],
                        columns=["Almond Butter","Peanut Butter","Cashew Butter"])
price_df
```

```python
# Shape aren't allinged, Let's Transpose
total_sales = price_df.dot(sales_amount.T)
total_sales
```
