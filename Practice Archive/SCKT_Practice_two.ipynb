```python
import pandas as pd 
import numpy as np 
import sklearn 
%matplotlib inline 
import matplotlib.pyplot as plt 
import datetime 
print(f"Last Updated: {datetime.datetime.now()}")
```

```python
car_sales = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-extended.csv")
car_sales.tail(5)
```

```python
X = car_sales.drop(["Price"],axis=1)
y = car_sales["Price"]
X.dtypes,y.dtypes
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

 Need to convert object data to numeric 
 Two ways to do that : 1. Using Sklearn's OneHotEncoder  2. Pandas dummies function 

```python
# SKlearn
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_features = ["Make","Colour","Doors"]

one_hot = OneHotEncoder()

transformer = ColumnTransformer([("Including Doors",one_hot,categorical_features)],remainder="passthrough")

transformed_X_in = transformer.fit_transform(X)

# Excluding Doors as an object datatype 
# Sklean 
categorical_variable = ["Make","Colour"]
transformer = ColumnTransformer([("Excluding Doors",one_hot,categorical_variable)],remainder="passthrough")
transformed_X_ex = transformer.fit_transform(X)

## Score 
X_train , X_test , y_train , y_test = train_test_split(transformed_X_in,y,test_size=0.4)
model = RandomForestRegressor()
model.fit(X_train,y_train)
score_in = model.score(X_test,y_test)

X_train, X_test , y_train , y_test = train_test_split(transformed_X_ex,y,test_size=0.4)
model.fit(X_train,y_train)
score_ex = model.score(X_test,y_test)

print(f"Score using OneHotEncoding (Doors Included) : {score_in*100:.2f}%")
print("")
print(f"Score using OneHotEncoding (Doors Excluded) : {score_ex*100:.2f}%")
```

```python
# Pandas 
# Including Doors
categorical_features = ["Make","Colour","Doors"]
car_sales["Doors"] = car_sales["Doors"].astype(object)
dummies_in = pd.get_dummies(data=car_sales[categorical_features],dtype=int)
X_train,X_test,y_train,y_test = train_test_split(dummies_in,y,test_size=0.4)
model = RandomForestRegressor()
model.fit(X_train,y_train)
dummies_score_in = model.score(X_test,y_test)

# Excluding Doors
categorical_variable = ["Make","Colour"]
dummies_ex = pd.get_dummies(data=car_sales[categorical_features],dtype=int)
X_train,X_test,y_train,y_test = train_test_split(dummies_ex,y,test_size=0.4)
model = RandomForestRegressor()
model.fit(X_train,y_train)
dummies_score_ex = model.score(X_test,y_test)

# Score
print(f"Score using pandas dummies (Doors Included): {dummies_score_in*100:.2f}%")
print("")
print(f"Score using pandas dummies (Doors Excluded): {dummies_score_ex*100:.2f}%")
```
