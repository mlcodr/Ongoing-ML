```python
import pandas as pd 
import numpy as np 
%matplotlib inline 
import matplotlib.pyplot as plt
import sklearn
import datetime
print(f"Using Sklearn Version: {sklearn.__version__}")
print(f"Last Updated: {datetime.datetime.now()}")
```

```python
heart_disease = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv")
heart_disease.tail()
```

```python
X = heart_disease.drop('target',axis=1)
y = heart_disease['target']
X.tail(3)
```

```python
y.head(5), y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X,y,test_size = 0.25,random_state=42)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.get_params()
```

```python
model.fit(X = X_train, y = y_train)
```

```python
y_preds = model.predict(X = X_test)
y_preds
```

```python
# Evaluate 
# Training Data 
train_acc = model.score(X = X_train, y = y_train)
print(f"Accuracy on the training dataset is: {train_acc*100:.2f}%")

# Testing Data 
test_acc = model.score(X = X_test, y = y_test)
print(f"Accuracy on the testing dataset is: {test_acc*100:.2f}%")

# More Evaluation Method 
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
classi_repo = classification_report(y_test,y_preds)
print(f"Classification Report of the model is: {classi_repo}")
confu_mat = confusion_matrix(y_test,y_preds)
print(f"Confusion Matrix of the model is: {confu_mat}")
acc_score = accuracy_score(y_test,y_preds)
print(f"Accuracy score of the model is: {acc_score*100:.2f}%")
```

```python
## Experimenting 
np.random.seed(42)
for n in range(200,475,25): 
    print(f"Trying with {n} estimators...")
    model = RandomForestClassifier(n_estimators = n).fit(X = X_train,y = y_train)
    print(f"Estimators accuracy after tuning hyperparameter --> {model.score(X = X_test,y = y_test)*100:.2f}%")

    ## Cross Validation 
    from sklearn.model_selection import cross_val_score
    cross_val_mean = np.mean(cross_val_score(model,X,y,cv = 10 ))
    print(f"Estimators accuracy after 10-fold cross-validation --> {cross_val_mean*100:.2f}%")
    
    print("")
```

```python
model.get_params()
```

```python
# Automatic Experiments using GridSearchCV
np.random.seed(42)
from sklearn.model_selection import GridSearchCV

param_grid = {"n_estimators" : [n for n in range(200,475,25)]}

grid = GridSearchCV(estimator = RandomForestClassifier(),
                   param_grid = param_grid, 
                   cv = 10,
                   verbose = 1) 
grid.fit(X,y)

print(f"The best hyperparameter values are: {grid.best_params_}")
print(f"With a values with --> {grid.best_score_ * 100 :.2f}% ")
```

```python
model = grid.best_estimator_
model
```

```python
print(f"Best score of the model on a single split dataset is : {grid.best_score_*100:.2f}%")
```

```python
## Saving model to a file 
import pickle
pickle.dump(model,open("random_forest_classifier_one","wb"))
```

```python
saved_model = pickle.load(open("random_forest_classifier_one","rb"))
print(f"Loaded Model on a single split dataset: {saved_model.score(X_test,y_test)*100:.2f}%")
```
