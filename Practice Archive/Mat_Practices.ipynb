```python
## Plot from a existing dataframe
car_sales = pd.read_csv(r"D:\Olympiad\Module 6\55 - car-sales.csv")
```

```python
# We will make a plot. x-axis will be the time_span and y-axis will be the product price increasing over years
car_sales['Price'] = car_sales["Price"].str.replace("[\$\,\.]","",regex = True)
```

```python
car_sales["Price"] = car_sales["Price"].str[:-1]
```

```python
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range("01/1/2025",periods=len(car_sales))
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
```

```python
car_sales
```

```python
car_sales.plot("Sales Date","Total Sales");
```

```python
## Scatter Plot from pandas 
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x = "Odometer (KM)", y = "Price", kind="scatter");
```
