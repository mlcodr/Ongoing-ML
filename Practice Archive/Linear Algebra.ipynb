## Solving Linear Systime 2 Variable 

A **system of linear equations** or **linear system** is a collection of one or more linear equations involving the same variable. For example: 
$$
\begin{cases}
-x_1 + 3x_2 = 7 ,\\
3x_1 + 2x_2 = 1
\end{cases}
$$
This is a system of two equations with two unknown variable **x_1,x_2**.To Solve a system of linear equations means to find such values of the variables **x_1,x_2** , that all of its equations are simultaneously satisfied.

A linear system is **inconsistent** if its has no solution and otherwise it is said to be **consistent**.
Consistent system can have one or infinite number of solutions.

```python
# In order to perform the calc we need to input seperately both co - efficiant and values (array).
# A stands for co-efficiant and b stands for value and x stands for answer 
A = np.array([[-1,3],
     [3,2]])
b = np.array([7,1])
x = np.linalg.solve(A,b) 
x
```

```python
x.shape,A.shape,b.shape
```

```python
## We need to checkout if the number or answer is correct 
x_1 , x_2 = x
check_1 = -1*x_1 + 3*x_2
check_2 = 3*x_1 + 2*x_2
print(f"Answer of the first equation is : {check_1}")
print(f"Answer of the second equation is : {check_2}")
```

```python
## The previous step was manual step now we'll see the automatic step to check the result 
x1,x2 = x
if np.allclose(np.dot(A,x),b): 
    print("Correct")
else: 
    print("Incorrect")
```

```python
## Checking if the equation have any answer or not by using np.linalg.det()
np.linalg.det(A)
```

### If np.linalg.det(A) == 0 that means it has no or infinite solution 
### If np.linalg.det(A) is (not=0) that means it has solution 

```python
d = np.linalg.det(A)
print(f"The determinant of A is: {d:.2f}")  ## .2f means there will be two decimal number.
```
