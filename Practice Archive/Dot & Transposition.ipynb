```python
import numpy as np
```

```python
np.random.seed(0)
mat_1 = np.random.randint(0,10, size=(5,3))
mat_1
```

```python
np.random.seed(0)
mat_2 = np.random.randint(0,10, size=(5,3))
mat_2
```

```python
## Element_wise Multiplication (Hadarmard Product)
mat_1 * mat_2
```

```python
# Dot Product
np.dot(mat_1,mat_2)
```
