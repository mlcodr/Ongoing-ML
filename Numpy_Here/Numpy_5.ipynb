```python
a2 = np.random.randint(0,10, size=(6))
a2
```

```python
"""
If you added up all the numbers in the array and divided by how many numbers there are.
"""
np.mean(a2)
```

```python
"""
The number that is exactly in the middle when you sort the array
(If you have an even count of numbers, it's the average of the two middle numbers)
"""
np.median(a2)
```

```python
np.min(a2)
```

```python
np.max(a2)
```

```python
"""
np.std() refers to the Standard Deviation

The typical or average distance each number is from the mean.
A higher number means the data is more spread out.
"""
np.std(a2)
```

```python
"""
np.var() refers to the SQUARED STANDARD DEVIATION

The standard deviation squared. It's another way to measure how spread out the data is,
but it's less intuitive than standard deviation because the units are squared.
"""
np.var(a2)
```

```python
# Standard Deviation  = Square Root of the Variance
np.sqrt(np.var(a2))
```

```python
# Variance  =  Square of the Standard Deviation
np.square(np.std(a2))
```
