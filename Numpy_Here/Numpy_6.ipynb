```python
# For Larger arrays the default value of a transpose is to swap the first and the last axes
# For example, (5,3,3) -> (3,3,5)
```

```python
m_1 = np.random.random(size = (2,3,4))
m_1
```

```python
m_1.shape, m_1.ndim
```

```python
trasnposed_m_1 = m_1.transpose()
trasnposed_m_1, trasnposed_m_1.shape
```

```python
trasnposed_m_1.shape == m_1.shape
```

```python
m_1.shape == trasnposed_m_1[::-1]  # -1 defines the reverseness of an array
```

```python
trasnposed_m_1 == np.swapaxes(m_1,0,1) # 0 , 1 means x and y axis sequencely
```
