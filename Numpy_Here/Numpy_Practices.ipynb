## Anatomy of a Numpy array
### Array - A list of numbers, can be multi-dimentional.
### Scalar - A single number(e.g:7).
### Vector - A list of numbers with 1-dimentional (e.g: np.array([1,2,3]))
### Matrix - A (Usually) multi-dimentional list of numbers (e.g:  array([[1. , 4. ],[2. , 5. ],[3.3, 6. ]]))
### Tensor - Tensor is above than Matrix as it is more than 3 dimentional or rank 3+

## Dot Product

```python
np.random.seed(0)
mat_1 = rng.integers(5,10,size=(4,5))
mat_1,mat_1.shape
```

```python
np.random.seed(0)
mat_2 = rng.integers(5,10,size=(5,3))
mat_2,mat_2.shape
```

```python

dot_1 = np.dot(mat_1,mat_2)
dot_1,dot_1.shape
```

```python
## The resulting shape is as same as the two outer matrix.
## For dot product the inner dimention of two arrays must match
## Can also achieve np.dot() with "@"  
## (however, they may behave differently at 3D+ arrays)
```

```python
print(mat_1 @ mat_2)
```

```python
np.random.seed(0)
mat_3 = rng.integers(5,10,size=(4,6))
mat_4 = rng.integers(4,10,size=(4,6))
mat_3, mat_4
```

```python
reshaped_mat_4 = mat_4.reshape(6,4)
```

```python
print(mat_3 @ reshaped_mat_4)
```

```python
# Another way to reshape.
reshaped_mat_4_1 = mat_4.swapaxes(1,0)
```

```python
mat_3 @ reshaped_mat_4_1 
```

```python
# We can also do transpose.
mat_3 @ mat_4.T
```

## Sorting Array

```python
## np.sort()
scores = np.array([80,78,79,94,87,82,80])
sorted_scores = np.sort(scores)
sorted_scores
```

```python
## np.argsort()
order_of_finisher = np.argsort(scores) ## This represents the position of each finisher according to the sorted array
order_of_finisher
```

```python
## np.argmax()
highest_scores = np.argmax(scores) ## This represents the highest scorers position
highest_scores
```

```python
## np.argmin()
lowest_scores = np.argmin(scores) ## This represents the lowest scorers position
lowest_scores
```
