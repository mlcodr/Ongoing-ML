# Manipulating and Comparing Arrays

## Arithmetic

```python
import numpy as np
```

```python
a1 = np.array([1,2,3])
a1
```

```python
one = np.ones(10)
one
```

```python
# Shape Must Be Simillar
a1 + one
```

```python
# Adding
ones = np.ones(3)
a1 + ones
```

```python
# Substracting
a1 - ones
```

```python
# Multiplying
a1 * ones
```

```python
a2 = np.array([2.5,2,6])
```

```python
a1 * a2
```

```python
a3 = np.random.random(size=(2,2))
```

```python
a4 = np.random.randint(0,10,size=(2,2))
```

```python
a3 * a4
```

```python
a4/a3
```

```python
# Floor Devision denotes as : //
# It removes the decimal Value
a4 // a3
```

```python
a4 ** 2
```

```python
# or we can use the square() Function --
np.square(a4)
```

```python
# Adding two arrays
np.add(a4,a3)
```

```python
a1 / 2
```

```python
# it shows us what remains after dividing a1 by 2  ( Only The Remainder )
a1 % 2
```
