## Just basic Comparison Operators

```python
import numpy as np
```

```python
array_1 = np.array([1,2,3])
array_2 = np.array([[1,2,3],
                  [4.,5.,6.5]])
```

```python
array_1 < array_2
```

```python
array_1 > array_2
```

```python
var_1 = array_1 >= array_2
```
