### What if there is some missing values in the data 
Holes in the data means holes in the pattern that your model is learning from 
Handling Missing Data (Imputation vs. Removal)

In machine learning, missing values (like NaN or blanks) can break your model. You generally have two ways to fix this:

    Imputation: Filling missing values with calculated data (e.g., the mean, median, or a constant).

    Removal: Deleting rows or columns containing missing values (simple, but reduces your dataset size).
